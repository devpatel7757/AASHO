```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/chicago_crimes_raw.csv')
print("Loaded! Shape:", df.shape)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p')

df['Year']    = df['Date'].dt.year
df['Month']   = df['Date'].dt.month
df['Day']     = df['Date'].dt.day
df['Hour']    = df['Date'].dt.hour
df['DayOfWeek'] = df['Date'].dt.day_name()
df['Season']  = df['Month'].map({
    12:'Winter',1:'Winter',2:'Winter',
    3:'Spring',4:'Spring',5:'Spring',
    6:'Summer',7:'Summer',8:'Summer',
    9:'Fall',10:'Fall',11:'Fall'
})
```

```python
print("Nulls before:", df.isnull().sum()[df.isnull().sum() > 0])

df.dropna(subset=['Latitude','Longitude','Primary Type'], inplace=True)

df['Location Description'] = df['Location Description'].fillna('Unknown')
df['Ward'] = df['Ward'].fillna(0).astype(int)
df['District'] = df['District'].fillna(0).astype(int)

print("Nulls after:", df.isnull().sum()[df.isnull().sum() > 0])
print("Final shape:", df.shape)
```
